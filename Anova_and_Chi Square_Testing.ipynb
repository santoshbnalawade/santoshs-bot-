```python
# Import the Libraries
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')

# Display the first few rows
print(iris.head())
```

```python
# Check if sepal length differs significantly across species
```

```python
# ========================= ANOVA TEST =========================
anova_result = stats.f_oneway(
    iris[iris['species'] == 'setosa']['sepal_length'],
    iris[iris['species'] == 'versicolor']['sepal_length'],
    iris[iris['species'] == 'virginica']['sepal_length']
)

print("\nANOVA Test Results for Sepal Length Across Species:")
print(f"F-Statistic: {anova_result.statistic}, P-value: {anova_result.pvalue}")
```

**ANOVA Test:**

- If p-value < 0.05, there is a statistically significant difference in sepal length across species.
- If p-value > 0.05, no significant difference in sepal length across species.

```python
# ========================= CHI-SQUARE TEST =========================
# Convert numerical variables into categorical by binning
iris['sepal_length_cat'] = pd.cut(iris['sepal_length'], bins=3, labels=["Short", "Medium", "Long"])

# Convert categorical labels to numbers
label_encoder = LabelEncoder()
iris['species_code'] = label_encoder.fit_transform(iris['species'])
iris['sepal_length_code'] = label_encoder.fit_transform(iris['sepal_length_cat'])

# Create a contingency table
contingency_table = pd.crosstab(iris['species_code'], iris['sepal_length_code'])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2_stat}, P-value: {p_value}")
```
